```python
import numpy as np

T = np.linspace(300,1000) # degrees K
t = T/1000;

# CH4     T = 298-1300K valid temperature range
A=-0.703029
B=108.4773
C=-42.52157
D=5.862788
E=0.678565
F=-76.84376
G=158.7163
H=-74.87310

Hf_29815_CH4 = -74.5 # kJ/mol
S_29815_CH4 = 188.66 # J/mol/K

dH_CH4 = A*t + B*t**2/2 + C*t**3/3 + D*t**4/4 - E/t + F - H
S_CH4 = (A*np.log(t) + B*t + C*t**2/2 + D*t**3/3 - E/(2*t**2) + G)

# N2O  298-1400 K valid temperature range
A = 27.67988
B =51.14898
C =-30.64454
D =6.847911
E =-0.157906
F =71.24934
G =238.6164
H =82.04824

Hf_29815_N2O = 82.05 # kJ/mol
S_29815_N2O = 219.96 # J/mol/K

dH_N2O = A*t + B*t**2/2 + C*t**3/3 + D*t**4/4 - E/t + F - H
S_N2O = (A*np.log(t) + B*t + C*t**2/2 + D*t**3/3 - E/(2*t**2) + G)

#  N2    500 - 2000 valid temperature range
A =  19.50583
B =   19.88705
C =   -8.598535
D =  1.369784
E =  0.527601
F =  -118.0089
G =   212.3900
H = 0
Hf_29815_N2 =0  #this is Hf kJ/mol.
S_29815_N2 = 191.61

dH_N2 = A*t + B*t**2/2 + C*t**3/3 + D*t**4/4 - E/t + F - H
S_N2 = (A*np.log(t) + B*t + C*t**2/2 + D*t**3/3 - E/(2*t**2) + G)


#  CH3OH   273. - 1000.K valid temperature range
A=-3.93019
B=217.4160
C=-285.830
D=157.572
E=-21.6902
F=-200.859
G=238.916
H=-238.679

Hf_29815_CH3OH = -200.8 # kJ/mol
S_29815_CH3OH = 1239.9 # J/mol/K


dH_CH3OH = A*t + B*t**2/2 + C*t**3/3 + D*t**4/4 - E/t + F - H
S_CH3OH = (A*np.log(t) + B*t + C*t**2/2 + D*t**3/3 - E/(2*t**2) + G)





Hrxn_29815 = Hf_29815_CH3OH + Hf_29815_N2 - Hf_29815_CH4 - Hf_29815_N2O
Srxn_29815 = S_29815_CH3OH + S_29815_N2 - S_29815_CH4- S_29815_N2O
Grxn_29815 = Hrxn_29815 - 298.15*(Srxn_29815)/1000;

print('deltaH = {0:1.2f}'.format(Hrxn_29815))
print('deltaG = {0:1.2f}'.format(Grxn_29815))


Hrxn = Hrxn_29815 + dH_CH3OH+ dH_N2 - dH_CH4 - dH_N2O
Grxn = Hrxn - T*(S_CH3OH + S_N2 - S_CH4 - S_N2O)/1000

import matplotlib.pyplot as plt
plt.figure(); plt.clf()
plt.plot(T,Grxn, label='$\Delta G_{rxn}$')
plt.plot(T,Hrxn, label='$\Delta H_{rxn}$')
plt.xlabel('Temperature / K')
plt.ylabel('(kJ/mol )')
plt.legend( loc='best')


R = 8.314# J/mol/K
K = np.exp(-1000*Grxn/R/T);

plt.figure()
plt.plot(T,K)
plt.xlim([300, 1000])
plt.xlabel('Temperature (K)')
plt.ylabel('Equilibrium constant')
```

```python
# read CSV file
df = pd.read_csv('Al-NMR.csv')

subset = df.loc[(df['shift1'] >= -100) & (df['shift1'] <= 200)]


# define x value
shift1 = subset['shift1']  

# define y value
intensity1 = subset['intensity1']  

# create plot
fig, ax = plt.subplots()
ax.plot(shift1 , intensity1,color='black',linewidth=2.5)
ax.set_ylim(-10, 20)



plt.xlabel('Chemical Shift',fontsize=14)
plt.ylabel('Intensity/a.u.',fontsize=14)
plt.show()
```
